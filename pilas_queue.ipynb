```
%%writefile Queue.cpp

#include <iostream>
#include <cstdlib>
#include <functional>
#include <queue>

using namespace std;

template <typename Tipo>

struct Nodo{
    Tipo elemento;
    Nodo<Tipo>* siguiente;
    Nodo<Tipo>* anterior;
    Nodo(Tipo e){elemento=e;siguiente=anterior=nullptr;};
};

template <typename Tipo>
class Queue{

    Nodo<Tipo>* inicio;

    Nodo<Tipo>* fin;

    long cantidad;

    public:

    Queue(){

        inicio=fin=nullptr;

        cantidad=0;        

    };

    Tipo front(){        

          if(inicio!=nullptr)

            return inicio->elemento;

    }

    void front(function<void(Tipo)> f){        

          if(inicio!=nullptr)

            f(inicio->elemento);

    }

     void enqueue(Tipo e) {//O(1)

      Nodo<Tipo>* nuevo = new Nodo<Tipo>(e);

      if (inicio == nullptr)inicio = fin=nuevo;                

    else
      { 
      fin->siguiente = nuevo;
       nuevo->anterior=fin;
      fin = fin->siguiente;
      }
      ++cantidad;
    }

    void dequeue(){
        if(inicio==nullptr)return;        
        Nodo<Tipo>* tmp=inicio;
        if (cantidad==1){                    
            inicio=fin=nullptr;            
        }
        else {        
            inicio=inicio->siguiente;
            inicio->anterior=nullptr;                        
        }
      delete tmp;
      --cantidad;
    }

    int size(){        
        return cantidad;
    }

    bool empty(){
         if(cantidad==0){
            return false;
            }
        else{
            return true ;
             }
    }
    };


class Persona{

  string nombre;

  short edad;

  public:

  Persona(string nombre,short edad):nombre(nombre),edad(edad){}

  string toString(){return nombre+" - "+to_string(edad)+"\n";}


};



void ejemplo(){

   Queue<Persona*> perso;  
    cout<<perso.empty()<<endl;
    perso.enqueue(new Persona("Alejandro",20));
    perso.enqueue(new Persona("Rodrigo",24));
    perso.enqueue(new Persona("Marcos",34)); 

    cout<<perso.front()->toString();
    cout<<perso.size()<<endl;
    cout<<perso.empty()<<endl;
    perso.dequeue();
    cout<<perso.front()->toString();
    cout<<perso.size()<<endl;;
    cout<<perso.empty();
};

int main(){

    ejemplo();

}
```

```
%%writefile Pila.cpp

#include <iostream>

#include <cstdlib>

#include <functional>


using namespace std;

template <typename Tipo>

struct Nodo{

    Tipo elemento;

    Nodo<Tipo>* siguiente;

    Nodo<Tipo>* anterior;

    Nodo(Tipo e){elemento=e;siguiente=anterior=nullptr;};

};
template <typename Tipo>

class Pila{

    Nodo<Tipo>* inicio;

    Nodo<Tipo>* fin;

    long cantidad;

    public:

    Pila(){

        inicio=fin=nullptr;

        cantidad=0;        

    };

    Tipo top(){        

          if(fin!=nullptr)

            return fin->elemento;

    }

    void top(function<void(Tipo)> f){        

          if(fin!=nullptr)

            f(fin->elemento);

    }

     void push(Tipo e) {//O(1)

      Nodo<Tipo>* nuevo = new Nodo<Tipo>(e);

      if (inicio == nullptr)

        inicio = fin=nuevo;                

    else

      { 
          
      fin->siguiente = nuevo;

       nuevo->anterior=fin;

      fin = fin->siguiente;

      }

      ++cantidad;

    }

    void pop(){

        if(inicio==nullptr)return;        

        Nodo<Tipo>* tmp=fin;

        if (cantidad==1){                    

            inicio=fin=nullptr;            

        }

        else {        

            fin=fin->anterior;

            fin->siguiente=nullptr;                        

        }

      delete tmp;

      --cantidad;

    }

    int size(){        

        return cantidad;

    }

    bool empty(){
        if(cantidad==0){
            return false;
            }
        else{
            return true ;
             }
    }

};


class Persona{

  string nombre;

  short edad;

  public:

  Persona(string nombre,short edad):nombre(nombre),edad(edad){}

  string toString(){return nombre+" - "+to_string(edad)+"\n";}
};



void ejemplo(){

   Pila<Persona*> perso;  
     
    cout<<perso.empty()<<endl;
    perso.push(new Persona("Alejandro",20));

    perso.push(new Persona("Rodrigo",24));

    perso.push(new Persona("Marcos",34)); 

    cout<<perso.top()->toString();
    cout<<perso.size()<<endl;
    perso.pop();
    cout<<perso.top()->toString();
    cout<<perso.size()<<endl;
    cout<<perso.empty();
    
};

int main(){

    ejemplo();

}
```
